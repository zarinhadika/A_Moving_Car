```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from xgboost import XGBClassifier
import shap
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
df=pd.read_csv("/content/data.csv")
df=df.drop(columns=[col for col in ['id','Unnamed: 32'] if col in df.columns] )

le=LabelEncoder()
df['diagnosis']=le.fit_transform(df['diagnosis'])

X=df.drop(columns=['diagnosis'])
y=df['diagnosis']
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=32,test_size=0.4)

model=XGBClassifier()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
print("Accuracy : " , accuracy_score(y_test,y_pred))
print("Precision : " , precision_score(y_test,y_pred))
print("Recall : " , recall_score(y_test,y_pred))
print("F1 Score : ",f1_score(y_test,y_pred))


explainer=shap.TreeExplainer(model)
shap_values=explainer.shap_values(X_test)
shap.summary_plot(shap_values,X_test)

sample_idx=8
shap.initjs()
shap.force_plot(explainer.expected_value,shap_values[sample_idx],X_test.iloc[sample_idx])
```
